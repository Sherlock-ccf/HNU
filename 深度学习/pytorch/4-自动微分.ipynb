```python
#d(scalar)/d(vector_c)=vector_r
#d(vector_c)/d(scalar)=vector_c
#d(vector_c)/d(vector_c)=matrix
#自动求导在函数指定值上求导
#原理：将代码分解成操作子->计算式转化为无环图
import torch
x=torch.arange(4.0);
x.requires_grad_(True)
y=2*torch.dot(x,x)
print(y)
y.backward()
print(x.grad)
#自动求导默认情况下会累积，需要清零
x.grad.zero_()
y=2*torch.dot(x,x)
y.backward()
print(x.grad)
```

```python
#detach方法：将一个函数(计算图)转化为其对应的数值
x.grad.zero_()
y=torch.dot(x,x)
u=y.detach()
print(u,y,sep=' ')
```
